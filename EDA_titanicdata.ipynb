Bivariate Analysis

```python
print(df['Sex'].value_counts())
```

```python
pd.crosstab(df["Survived"],df['Sex']).plot(kind = 'bar')
plt.title(" survivor rate")
plt.show()
```

```python
pd.crosstab(df["Survived"],df['Pclass']).plot(kind = 'bar')
plt.title(" survivor rate")
plt.show()
```

```python
pd.crosstab(df["Fare"],df['Age']).plot(kind = 'bar')
plt.title(" survivor rate")
plt.show()
```

```python
corr = df.corr(numeric_only = True)
corr
```

```python
plt.imshow(corr,cmap = "coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)),corr.columns,rotation = 90)
plt.yticks(range(len(corr.columns)),corr.columns, rotation = 90)
plt.title("correlation matrix")
plt.show()
```

```python
plt.imshow(corr,cmap = "pink")
plt.colorbar()
plt.xticks(range(len(corr.columns)),corr.columns,rotation = 400)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.title("correlation matrix")
plt.show()
```
